```python
print("03_fahrenheit_to_celsius")

def temp():
  print("This is for converting fahrenheit to celsius!")
  fahrenheit = float(input("Enter temperature in Fahrenheit: "))
  celsius =  (fahrenheit - 32) * 5.0/9.0
  print(f"Temperature : {fahrenheit}F =  {celsius}C")
  # or
  # print(f"Temperature : {fahrenheit}F =  {celsius:.2f}C")

if __name__ == "__main__":
  temp()
```

```python
print("05_triangle_perimeter.")

def triangle():
  print("This is about sum of triangle sides.")
  side1:float = float(input("Enter the length of the side 1 :"))
  side2:float = float(input("Enter the length of the side 2 :"))
  side3:float = float(input("Enter the length of the side 3 :"))

  perimeter = side1 + side2 + side3

  print(f"The perimter of the triangle is {perimeter}")


if __name__ == "__main__":
    triangle()
```
