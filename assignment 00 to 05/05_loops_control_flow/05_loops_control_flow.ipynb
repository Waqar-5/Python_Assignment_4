```python
import random
print("00_guess_my_number.")

def main():
  secret_number = random.randint(1, 100)
  print("I am thinking of  number between 1 and 100....")

  guess = int(input("Enter a guess: "))

  while guess != secret_number:
    if guess < secret_number:
      print("Your guess is too low.")
    else:
      print("Your guess is too high.")
    guess = int(input("Enter a guess: "))
  # print("You got it!")
  print(f"Congratualation! The number was {secret_number}.")

if __name__ == "__main__":
  main()
```

```python
print("01_fibonacci.")

max_value = 10000

def main():
  a,b = 0,1
  print(a,b, end = " ")

  while True:
    c = a+b
    if c >= max_value:
      break
    print(c, end=" ")
    a,b = b,c

if __name__ == "__main__":
  main()
```

```python
print("02_print_events.")

def main():
  for i in range(20):
    print(i*2)

if __name__ == "__main__":
  main()
```

```python
print("03_wholesome_machine.")

correct_affirmation = "I am capable of doing anything. I put my mind too."

def main():
  print("Welcome to the Wholesome Machine.")
  while True:
    user_input = input("Please, type the following affermation: "+ correct_affirmation)
    if user_input == correct_affirmation:
      print("That\'s right!")
      break
    else:
      print("Hmmm! That was not the affirmation. Please Try Again!")

if __name__ == "__main__":
  main()
```
